```python
# Importar librerías
import pandas as pd
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

# Crear el DataFrame
data = {
    'Producto A': [2.5, 1.5, 3.5, 5.5, 2.0],
    'Producto B': [3.5, 2.5, 4.0, 5.0, 3.0],
    'Producto C': [3.0, 2.0, 4.5, 5.5, 2.5],
    'Producto D': [4.5, 3.5, 5.0, 6.0, 3.5]
}
df = pd.DataFrame(data, index=['R1', 'R2', 'R3', 'R4', 'R5'])

# Aplicar K-Means
kmeans = KMeans(n_clusters=2, random_state=0)
df['Cluster'] = kmeans.fit_predict(df)

# Visualizar con seaborn (usando solo 2 productos para simplificar)
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Producto A', y='Producto D', hue='Cluster', palette='Set1', s=100)
plt.title('Visualización de Clústeres con K-Means')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Crear el dataset manualmente (como se muestra en el archivo 3.13)
data = {
    'Producto A': [2.5, 1.5, 3.5, 5.5, 2.0],
    'Producto B': [3.5, 2.5, 4.0, 5.0, 3.0],
    'Producto C': [3.0, 2.0, 4.5, 5.5, 2.5],
    'Producto D': [4.5, 3.5, 5.0, 6.0, 3.5]
}

regiones = ['R1', 'R2', 'R3', 'R4', 'R5']
df = pd.DataFrame(data, index=regiones)

print("📊 Datos originales:")
print(df)

# 2. Aplicar K-means con 2 clústeres (puede ajustarse)
kmeans = KMeans(n_clusters=2, random_state=42)
df['Cluster'] = kmeans.fit_predict(df)

print("\n📌 Resultados con asignación de clúster:")
print(df)

# 3. Visualizar los clústeres
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='Producto A', y='Producto B',
    hue='Cluster', palette='Set1', data=df, s=100
)
plt.title("Clustering de Regiones basado en Productos A y B")
plt.xlabel("Producto A")
plt.ylabel("Producto B")
plt.grid(True)
plt.show()
```

```python
# Paso 1: Librerías necesarias
import pandas as pd
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

# Paso 2: Crear el conjunto de datos
data = {
    'Producto A': [2.5, 1.5, 3.5, 5.5, 2.0],
    'Producto B': [3.5, 2.5, 4.0, 5.0, 3.0],
    'Producto C': [3.0, 2.0, 4.5, 5.5, 2.5],
    'Producto D': [4.5, 3.5, 5.0, 6.0, 3.5],
}
df = pd.DataFrame(data, index=['R1', 'R2', 'R3', 'R4', 'R5'])

# Paso 3: Aplicar KMeans (K=2)
kmeans = KMeans(n_clusters=2, random_state=0)
df['Cluster'] = kmeans.fit_predict(df)

# Paso 4: Mostrar la tabla resultante
print("📋 Datos con Clústeres Asignados:")
display(df)

# Paso 5: Visualización con Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='Producto A',
    y='Producto B',
    hue='Cluster',
    data=df,
    palette='Set2',
    s=100
)
plt.title('Visualización de Clústeres con K-Means')
plt.xlabel('Producto A')
plt.ylabel('Producto B')
plt.grid(True)
plt.show()
```
